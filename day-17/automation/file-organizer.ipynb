```python
import os
import os.path
import shutil
import subprocess
```

```python
# Get the current working directory
os.getcwd()
```

```python
# Change the directory to mydir
os.chdir(r"./mydir")
```

```python
os.getcwd()
```

```python
# List the directory
files = os.listdir()
files
```

```python
f = 'road.jpg'
os.path.splitext(f)[1][1:]
```

```python
dirs = set([os.path.splitext(f)[1][1:] for f in files])
dirs 
```

```python
# Create the directories
for dir in dirs:
    os.mkdir(dir)
```

```python
# Move the files
for file in files:
    src = file
    dest = os.path.splitext(file)[1][1:]
    finalDest = os.path.join(dest, file)
    shutil.move(src, finalDest)
    # print(finalDest)
```
